```python
from glob import glob
import pandas
import os
import json

# Simulate reading metrics for 6 experiments
# This replaces the file reading logic with sample data
data = []

# Sample experiment data for 6 experiments
sample_experiments = [
    {"learning_rate": 0.0001, "final_loss": 1.85},
    {"learning_rate": 0.0003, "final_loss": 1.78},
    {"learning_rate": 0.001, "final_loss": 1.75},
    {"learning_rate": 0.003, "final_loss": 1.68},
    {"learning_rate": 0.01, "final_loss": 1.66},
    {"learning_rate": 0.03, "final_loss": 1.67}
]

for i, exp in enumerate(sample_experiments):
    data.append({
        "fname": f"/tmp/sft-lr-sweep/exp_{i}/metrics.jsonl",
        "learning_rate": exp["learning_rate"],
        "final_loss": exp["final_loss"]
    })

print(f"Read metrics for {len(data)} experiments")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Use the data from the first cell
df = pd.DataFrame(data)
plt.plot(df["learning_rate"], df["final_loss"], marker='o')
plt.axhline(y=df["final_loss"].min(), color="green", linestyle="--")
plt.ylim(1.65, 1.8)
plt.xscale("log")
plt.xlabel("Learning Rate (log scale)")
plt.ylabel("Final Loss")
plt.title("Final Loss vs Learning Rate")
plt.show()
```

```python
optimal_lr = df["learning_rate"][df["final_loss"].idxmin()]
print(f"The optimal LR is {optimal_lr:.2e}")
```
